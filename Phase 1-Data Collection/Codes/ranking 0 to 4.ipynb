```python
def make_list(ind): # make list of people by index of columns
        name=meta.iloc[:,ind]
        Name_list = name.tolist() 
        s=pd.Series(Name_list)
        a=s.str.split(pat="\n")
        for i in range(1,meta.shape[0]):
             if type(a[i])!=float:
           
                    a[0]+=a[i]
            
        A = list(dict.fromkeys(a[0]))
#         print(A)
        return A
```

```python
prod=col_list.index('Producer(s)')
dire=col_list.index('Director(s)')
writer=col_list.index('Writer(s)')
star=col_list.index('Stars')
dist=col_list.index('Distributor')
print(prod,dire,writer,star,dist)
```

rank each number between 0 to 4

```python
def checK_rank(num,sep):
    if num<=sep:
        return 0
    if sep<num and num<=(2*sep):
        return 1
    if (2*sep)< num and num<=(3*sep):
        return 2
    if (3*sep)< num and num<=(4*sep):
        return 3
    if (4*sep)< num and num<=(5*sep):
        return 4
```

make new data by for people in movies IMDB votes*IMDB score 

```python
def ranking(data,x,ind):
        A= make_list(ind) # A is list of names
        df=data
        additive_rank_list=np.zeros(df.shape[0])
        rank_list=np.zeros(df.shape[0])
        for pd in A:
            for i in range (df.shape[0]):
                  if type(df[x][i])!=float:     # if it isn't Nan
                    if pd in df[x][i]: # if that name in row i
                                additive_rank_list[i]+=df['IMDB votes'][i]*df['IMDB score'][i]
#                                 print(rank_list[i])
        
        MAX=max(additive_rank_list) 
        sep=MAX/5
        for i in range(len(additive_rank_list)):
                 rank_list[i]=checK_rank(additive_rank_list[i],sep)                    
                  
        df['%s rank'%x]=rank_list
        df['additive %s rank'%x]=additive_rank_list
#         df.to_csv('data frame with %s ranking'%x)
        return 'done'
```

```python
%%time
ranking(meta,col_list[prod],prod)
ranking(meta,col_list[dire],dire)
ranking(meta,col_list[writer],writer)
ranking(meta,col_list[star],star)
ranking(meta,col_list[dist],dist)
```
